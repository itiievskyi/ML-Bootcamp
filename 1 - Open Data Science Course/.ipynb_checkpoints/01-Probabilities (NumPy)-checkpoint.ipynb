```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
people = np.array([[168,59], [191,90], [175,70], [171,65], [180,100], [155,55], [165,60], [174,64], [187,82], [184,76], [166,55], [162,50], [176,70], [170,68], [168,57]])
```

```python
people
```

```python
x = people[:,0]
y = people[:,1]
```

```python
np.median(people, axis=0)
```

```python
np.mean(people, axis=0)
```

```python
np.var(people, axis=0)
```

```python
np.std(people, axis=0)
```

```python
A = np.vstack([x, np.ones(len(x))]).T
```

```python
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
```

```python
plt.figure(num=1, figsize=(8, 6), dpi=200)

plt.plot(x, m*x + c, 'r', label='Fitted line')
plt.scatter(x=people[:,0], y=people[:,1])
plt.show()
```
